# Ejercicio 2: Incendios en México

```python
df= pd.read_csv("datos/Serie_historica_anual_incendios_2017.csv")
df.head()
```

<div class="alert alert-success">
    
1. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para los 5 estados más afectados 
    
</div> 

```python
df.columns
```

```python
total_hectareas_top_5_estados = df.groupby('Estado')['Total Hectareas'].sum().sort_values(ascending = False).head()
```

```python
#Estilo Matlab
plt.figure(figsize=(6,4))  

plt.bar(total_hectareas_top_5_estados.index, total_hectareas_top_5_estados.values, color = 'blue')          

plt.title('Total de hectáreas afectadas en los 5 estados mas afectados',fontsize=10)
plt.xlabel('Estados',fontsize=10)
plt.ylabel('Total de hectáreas afectadas',fontsize=10);
plt.show()
```

```python
#Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(6,4), dpi=100)

ax.bar(total_hectareas_top_5_estados.index, total_hectareas_top_5_estados.values, color = 'red')          
ax.set(title='Total de hectáreas afectadas en los 5 estados mas afectados', xlabel= 'Estados' , 
       ylabel='Total de hectáreas')

plt.show()
```

<div class="alert alert-success">
    
2. Utilizar una grafica de barras para comparar el total de hectáreas afectadas para Yucatán para los meses: abril, mayo, junio, julio y agosto 
    
</div> 

```python
meses_seleccionados = ((df['Mes Acontecimiento'] == 'abril') | (df['Mes Acontecimiento'] == 'mayo') | 
                       (df['Mes Acontecimiento'] == 'junio') | (df['Mes Acontecimiento'] == 'julio') | 
                       (df['Mes Acontecimiento'] == 'agosto'))

df_yucatan = df[(df['Estado'] == 'Yucatán') & meses_seleccionados].groupby('Mes Acontecimiento')['Total Hectareas'].sum().reset_index()

meses = {'abril':0, 'mayo':0, 'junio':0, 'julio':0, 'agosto':0}

for mes in df_yucatan.values:
    if mes[0] in meses:
        meses[mes[0]] = mes[1]
```

```python
#Estilo Matlab
plt.figure(figsize=(4,3))  

plt.bar(meses.keys(), meses.values(), color = 'blue')        

plt.title('Total de hectáreas afectadas en Yucatán',fontsize=10)
plt.xlabel('Meses',fontsize=10)
plt.ylabel('Hectáreas',fontsize=10);
plt.show()
```

```python
#Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(4,3), dpi=100)

ax.bar(meses.keys(), meses.values(), color = 'red')          
ax.set(title='Total de hectáreas afectadas en Yucatán', xlabel= 'Meses' , 
       ylabel='Hectáreas')

plt.show()
```

# Ejercicio 3: Pacientes confirmados con Covid en México


Fuente: https://datos.covid-19.conacyt.mx/#DownZCSV

```python
df= pd.read_csv("datos/Casos_Diarios_Estado_Nacional_Confirmados_20201129.csv")
df.head()
```

<div class="alert alert-success">
    
1. Hacer una grafica de barras comparando todos los meses, considerando el total nacional (por mes) 
    
</div> 

```python
df=df.loc[:,'nombre':]
df.head()
```

```python
df = df.set_index('nombre').T
df.head()
```

```python
df=df.reset_index()
df.head()
```

```python
df = df.rename(columns = {'index': 'Fecha'})
df.head()
```

```python
df['Fecha'] = pd.to_datetime(df['Fecha'], format= '%d-%m-%Y')
df = df.set_index('Fecha')
df.head()
```

```python
total_nacional_por_mes = df['Nacional'].resample('M').sum()
```

```python
#Estilo Matlab
plt.figure(figsize=(12,5))  

plt.bar(total_nacional_por_mes.index.month_name(), total_nacional_por_mes.values, color = 'blue')          

plt.title('Total nacional por mes',fontsize=10)
plt.xlabel('Meses',fontsize=10)
plt.ylabel('Pacientes confirmados',fontsize=10);
plt.show()
```

```python
#Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(12, 5), dpi=100)

ax.bar(total_nacional_por_mes.index.month_name(), total_nacional_por_mes.values, color = 'red')          
ax.set(title='Total nacional por mes', xlabel= 'Meses' , 
       ylabel='Pacientes confirmados')

plt.show()
```

<div class="alert alert-success">
    
2. Hacer una grafica de barras comparando  los meses de agosto, septiembre, octubre y noviembre, considerando solo las entidades de Yucatán, Distrito Federal  y Jalisco
    
</div> 

```python
df_estados_seleccionados = df[['YUCATAN', 'DISTRITO FEDERAL', 'JALISCO']]
df_estados_seleccionados.tail(8)
```

```python
df_estados_seleccionados = df_estados_seleccionados.copy()
df_estados_seleccionados.loc[:, 'Total'] = df_estados_seleccionados.sum(axis = 1)
df_estados_seleccionados.tail(8)
```

```python
df_estados_seleccionados = df_estados_seleccionados[df_estados_seleccionados.index.month.isin([8, 9, 10, 11])]
```

```python
total_meses_estados_seleccionados = df_estados_seleccionados['Total'].resample('M').sum()
```

```python
#Estilo Matlab
plt.figure(figsize=(6,4))  

plt.bar(total_meses_estados_seleccionados.index.month_name(), total_meses_estados_seleccionados.values, color = 'blue')          

plt.title('Total por mes',fontsize=10)
plt.xlabel('Meses',fontsize=10)
plt.ylabel('Pacientes confirmados',fontsize=10);
plt.show()
```

```python
#Estilo orientado a objetos
fig, ax = plt.subplots(figsize=(6, 4), dpi=100)

ax.bar(total_meses_estados_seleccionados.index.month_name(), total_meses_estados_seleccionados.values, color = 'red')          
ax.set(title='Total por mes', xlabel= 'Meses' , 
       ylabel='Pacientes confirmados')

plt.show()
```
