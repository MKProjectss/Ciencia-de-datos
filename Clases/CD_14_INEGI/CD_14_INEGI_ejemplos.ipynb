```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import requests
import json
# 8f370b20-48bb-96fa-c562-f2b70429fa77
```

```python
#Llamado al API
url= 'https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000002/es/0700/false/BISE/2.0/8f370b20-48bb-96fa-c562-f2b70429fa77?type=json'
respuesta = requests.get(url)
respuesta.status_code
```

```python
url= 'https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000002/es/0700/false/BISE/2.0/8f370b20-48bb-96fa-c562-f2b70429fa77?type=json'
respuesta = requests.get(url)
if respuesta.status_code == 200:
    #print(respuesta.content)
    contenido = json.loads(respuesta.content)

contenido
```

```python
datos_hombres = contenido['Series'][0]['OBSERVATIONS']
```

```python
poblacion_hombres = []
periodo = []

for dato in datos_hombres:
    poblacion_hombres.append(int(float(dato['OBS_VALUE'])))
    periodo.append(int(dato['TIME_PERIOD']))

print(poblacion_hombres)
print(periodo)
```

```python
url= 'https://www.inegi.org.mx/app/api/indicadores/desarrolladores/jsonxml/INDICATOR/1002000003/es/0700/false/BISE/2.0/8f370b20-48bb-96fa-c562-f2b70429fa77?type=json'
respuesta = requests.get(url)
if respuesta.status_code == 200:
    #print(respuesta.content)
    contenido = json.loads(respuesta.content)
    
datos_mujeres = contenido['Series'][0]['OBSERVATIONS']

poblacion_mujeres = []
periodo = []

for dato in datos_mujeres:
    poblacion_mujeres.append(int(float(dato['OBS_VALUE'])))
    periodo.append(int(dato['TIME_PERIOD']))

print(poblacion_mujeres)
print(periodo)
```

```python
df_poblacion = pd.DataFrame()
df_poblacion['periodo'] = periodo
df_poblacion['hombres'] = poblacion_hombres
df_poblacion['mujeres'] = poblacion_mujeres
df_poblacion['total'] = df_poblacion['hombres'] + df_poblacion['mujeres']
df_poblacion
```

```python
plt.plot(df_poblacion['periodo'], df_poblacion['total'], 'k-', label = 'Total')
plt.plot(df_poblacion['periodo'], df_poblacion['mujeres'], 'r-.', label = 'Mujeres')
plt.plot(df_poblacion['periodo'], df_poblacion['hombres'], 'b-.', label = 'Hombres')
plt.show()
```

```python
df = pd.melt(df_poblacion, id_vars='periodo',
            var_name='Categoria', value_name='Total')
df
```

```python
df = pd.melt(df_poblacion, id_vars='periodo',
            var_name='Categoria', value_name='Poblacion')
df
```

```python
import seaborn as sns
plt.figure(figsize=(15,8))
plt.title('Population growth in Mexico')
sns.barplot(x='periodo', y='Poblacion', data=df, hue='Categoria', palette=["#66053A","#0B50D3","#D30B57"])
plt.gca().legend().set_title('Gender')
plt.xticks(rotation=90)
plt.ylabel("Population (in hundreds of millions)")
plt.show()
```
