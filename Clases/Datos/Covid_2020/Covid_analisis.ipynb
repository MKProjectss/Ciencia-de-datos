```python
# Encontrar número de pacientes mujeres mayores de 50 años de nacionalidad MEXICANA
len(df[(df['SEXO'] == 1) & (df['EDAD'] > 50) & (df['NACIONALIDAD'] == 1)])
```

```python
# Numero de personas que fueron intubados, tienen tabaquismo y son mayores a 60 años
len(df[(df['INTUBADO'] == 1) & (df['TABAQUISMO'] == 1) & (df['EDAD'] > 60)])
```

```python
# ordenar por nacionalidad
def etiqueta_nacionalidad(nacionalidad):
    if nacionalidad == 1:
        return 'MEXICANA'
    elif nacionalidad == 2:
        return 'EXTRANJERA'
    elif nacionalidad == 99:
        return 'NO ESPECIFICADO'
    
df['NACIONALIDAD_ETIQUETA']= df['NACIONALIDAD'].apply(etiqueta_nacionalidad)
df.sort_values(['NACIONALIDAD_ETIQUETA', 'PAIS_NACIONALIDAD'])
```
